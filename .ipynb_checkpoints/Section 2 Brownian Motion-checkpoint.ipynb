```python
from matplotlib import pyplot as plt
import numpy as np
import random
import math

t = 1
n = 500
dt = t/n

dW = []
w = []

dW.append(math.sqrt(dt)*random.gauss(0, 1))
w.append(dW[0])
for i in range(n-1):
    dW.append(math.sqrt(dt)*random.gauss(0, 1))
    w.append(w[i]+dW[i+1])
times = [i/n for i, _ in enumerate(w)]

plt.plot(times, w, color = 'green', linestyle = 'solid')
plt.title("Brownian Path Simulation")
plt.ylabel("Relative Position")
plt.xlabel("Time")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import random
import math

t = 1
n = 500
dt = t/n

dw = np.sqrt(dt)*np.random.randn(1,n-1)
w = np.cumsum(dw)
times = [i/n for i, _ in enumerate(w)]

plt.plot(times, w, color = 'green', linestyle = 'solid')
plt.title("Brownian Path Simulation")
plt.ylabel("Relative Position")
plt.xlabel("Time")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import random
import math

t = 1
n = 500
dt = t/n
m = 1000

dw = np.sqrt(dt)*np.random.randn(m,n-1)
w = np.zeros([m,n])
w[:,1:n] = np.cumsum(dw, axis=1)
times = [i/n for i, _ in enumerate(w[0])]
u = np.exp(np.tile(times,(m,1)) + 0.5*w);
for k in range(5):
    plt.plot(times, np.reshape(u[k, :], n))
mean = np.mean(u, axis=0)
plt.plot(times, mean, color = 'black', linestyle = 'solid')
plt.title("Brownian Path Simulation", fontsize = 'x-large')
plt.ylabel(r"$U(W(t))=e^{t+\frac{1}{2}W(t)}$", fontsize = 'x-large')
plt.xlabel("Time", fontsize = 'x-large')
plt.show()
```
