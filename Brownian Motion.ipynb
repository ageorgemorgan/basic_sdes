```python
from matplotlib import pyplot as plt
import random
import math

t = 1
n = 500
dt = t/n

dW = []
w = []

dW.append(math.sqrt(dt)*random.gauss(0, 1))
w.append(dW[0])
for i in range(n-1):
    dW.append(math.sqrt(dt)*random.gauss(0, 1))
    w.append(w[i]+dW[i+1])
times = [i for i, _ in enumerate(w)]

plt.plot(times, w, color = 'green', linestyle = 'solid')
plt.title("Brownian Path Simulation")
plt.ylabel("Relative Position")
plt.xlabel("Time")
plt.show()
```

```python
from matplotlib import pyplot as plt
import random
import math

t = 1
n = 500
dt = t/n
m = 1000


dW = []
w = []
u = []

for i in range(m):
    dW.append([])
    w.append([])
    u.append([])

for i in range(m):
    dW[i].append(math.sqrt(dt)*random.gauss(0, 1))
    w[i].append(dW[i][0])
    u[i].append(math.e**(0.5*w[i][0]))
    for j in range(n-1):
        dW[i].append(math.sqrt(dt)*random.gauss(0, 1))
        w[i].append(w[i][j]+dW[i][j+1])
        u[i].append(math.e**(float(j+1)/n+0.5*w[i][j+1]))

mean = []

for i in range(n):
    mean.append(0)
    for j in range(m):
        mean[i] += u[j][i]
    mean[i] = float(mean[i])/m

times = [i for i, _ in enumerate(w[0])]
plt.plot(times, u[0], color = 'red', linestyle = 'solid')
plt.plot(times, u[1], color = 'orange', linestyle = 'solid')
plt.plot(times, u[2], color = 'cyan', linestyle = 'solid')
plt.plot(times, u[3], color = 'green', linestyle = 'solid')
plt.plot(times, u[4], color = 'blue', linestyle = 'solid')
plt.plot(times, u[5], color = 'purple', linestyle = 'solid')
plt.plot(times, mean, color = 'black', linestyle = 'solid')
plt.title("Brownian Path Simulation")
plt.ylabel("Relative Position")
plt.xlabel("Time")
plt.show()
```
